```python
import os
import cv2
import shutil
```

```python
from pathlib import Path
```

```python
ROOT_DIRECTORY = r"C:\DEV\IAN\avocado-gis\neural\original-images"
DESTINATION_DIRECTORY = r"C:\DEV\IAN\avocado-gis\neural\working-images"
OUTPUT_IMAGE_SIZE = 224
```

```python
def treat_image(path):
    image = cv2.imread(path)
    image = cv2.resize(image, (OUTPUT_IMAGE_SIZE, OUTPUT_IMAGE_SIZE))
    return image
```

```python
def save_image(image, path, name):
    full_path = os.path.join(path, name)
    print(f"Saving treated image to: {full_path}")
    cv2.imwrite(full_path, image)
```

```python
def process_images(reference_directory, destination_path):
    if os.path.exists(destination_path):
        shutil.rmtree(destination_path)
    os.makedirs(destination_path)

    for root, dirs, files in os.walk(reference_directory):
        for file in files:
            full_path = Path(os.path.join(root, file))
            class_name = os.path.basename(Path(full_path).resolve().parents[0])
            purpose_type = os.path.basename(Path(full_path).resolve().parents[1])

            treated_image = treat_image(str(full_path))

            subdirectory = os.path.join(purpose_type, class_name)
            new_path = os.path.join(destination_path, subdirectory)

            if not os.path.exists(new_path):
                os.makedirs(new_path)

            print(f'Destination: {new_path}')
            save_image(treated_image, new_path, file)
```

```python
process_images(ROOT_DIRECTORY, DESTINATION_DIRECTORY)
```
